```python
test=folium.Map(location=[0.00,0.00], zoom_start=1)
```

```python


places=dataset[['latitude','longitude']]
```

```python
place=places.values.tolist()
```

```python
def y(point,i,color):
    str1="""{}
    Cases: {}
    Deaths: {}
    Recovered: {}""".format(dataset['location'][i],dataset['confirmed'][i],dataset['dead'][i],dataset['recovered'][i])
    folium.Marker(location=point,
                  popup=str1,
                  icon=folium.Icon(color=color, icon='tint', icon_color='white')
    ).add_to(test)
```

```python
i=0
for point in place:
    if dataset['confirmed'][i]>4000:
        y(point,i,'red')
    elif dataset['confirmed'][i]>2500:
        y(point,i,'orange')
    elif dataset['confirmed'][i]>1500:
        y(point,i,'blue')
    else:
        y(point,i,'green')
        
    i+=1
```
